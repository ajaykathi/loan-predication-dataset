```python
#create the pieplot
data=df.groupby('Gender')['Education'].count()
explode=[0.1]*len(data)
plt.figure(figsize=(6,6))
plt.pie(data,
        labels=data.index,
        explode=explode,
        autopct='%1.1f%%',
         shadow=True,
       startangle=140)
plt.title('Gender wish Education ')
plt.show()
```

```python
data=df.groupby('Property_Area')['ApplicantIncome'].count()
explode=[0.1]*len(data)
plt.figure(figsize=(6,6))
plt.pie(data,
        labels=data.index,
        explode=explode,
        autopct='%1.1f%%',
         shadow=True,
       startangle=140)
plt.title('ApplicantIncome wish Property_Area')
plt.show()
```

```python
data=df.groupby('Gender')['LoanAmount'].count()
explode=[0.1]*len(data)
plt.figure(figsize=(6,6))
plt.pie(data,
        labels=data.index,
        explode=explode,
        autopct='%1.1f%%',
         shadow=True,
       startangle=140)
plt.title('LoanAmount wish Gender')
plt.show()
```

```python
#create the barplot
data=df.groupby('Married')['LoanAmount'].sum()
colors=['red','green']
plt.bar(data.index,data.values,color=colors)
for i in range(len(data.index)):
    plt.text(i,data.values[i],str(data.values[i]),ha='center')
plt.title('Married wise LoanAmount')
plt.xlabel("Married")
plt.ylabel("LoanAmount")
plt.show()
```

```python
data=df.groupby('Education')['LoanAmount'].sum()
colors=['blue','orange']
plt.bar(data.index,data.values,color=colors)
for i in range(len(data.index)):
    plt.text(i,data.values[i],str(data.values[i]),ha='center')
plt.title('Education wise LoanAmount')
plt.xlabel("Education")
plt.ylabel("LoanAmount")
plt.show()
```

```python
data=df.groupby('Property_Area')['Loan_Amount_Term'].sum()
colors=['skyblue','pink','lightgreen']
plt.bar(data.index,data.values,color=colors)
for i in range(len(data.index)):
    plt.text(i,data.values[i],str(data.values[i]),ha='center')
plt.title('Property_Area wish Loan_Amount_Term')
plt.xlabel("Property_Area")
plt.ylabel("Loan_Amount_Term")
plt.show()
```
